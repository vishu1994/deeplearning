## Pyplot
#### 1. Draw a line from position (0,0) to (6,250)

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])
plt.figure(figsize=(5,5))
plt.plot(xpoints, ypoints)
plt.show()
```

#### Draw only x,y points instead of a line
##### To plot only the markers, we can use shortcut string notation parameter 'o', which means 'rings'.

```python
xpoints = np.array([0,4,6,8,10])
ypoints = np.array([0, 6, 3, 16, 2])

plt.plot(xpoints, ypoints , 'o')
plt.show()
```

### Labels and Titles

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")
plt.title("Sports Watch Data")
plt.figure(figsize=(10,2))
print(plt.plot(x,y))
```

#### Setting Font properties for Title and Labels

```python
font1 = {
    'family' : 'serif',
    'color' : 'blue',
    'size' : 20
}
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])


font2 = {'family':'serif','color':'darkred','size':15}
plt.title("Sports Watch Data", fontdict = font1 , loc='left')
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)
plt.figure(figsize=(5,5))
plt.plot(x, y)
plt.show()
```
